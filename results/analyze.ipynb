```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
data = pd.read_csv('qbert_reproduction.csv')
```

```python
data
```

```python
x = data['Step']
y = data['true-galaxy-28 - ep/return']

y = scipy.signal.savgol_filter(y, 101, 1)
```

```python
plt.style.use('seaborn-v0_8-colorblind')
plt.plot(x, y)
plt.title('Qbert Reproduction')
plt.xlabel('Episode')
plt.ylabel('Return')
plt.savefig('qbert_reproduction.png', dpi=600)
plt.savefig('qbert_reproduction.pdf')
```
